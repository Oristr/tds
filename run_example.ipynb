## Credit Card
This dataset is a tabular dataset about credit card transactions in which a minor percentage of them are labeled as fraudulent and there rest are normal.

```python
X_train, y_train, X_test, y_test = load_data('credit-card')
best_model, results = automate_model_selection(X_train[:10000], y_train[:10000]) #to run faster
```

As we can see from the results, the best model is HBOS - Histogram-based Outlier Detection

```python
plot_results_dict(best_model, results)
```

## Loans
This dataset is a tabular dataset about loan requests, given a request we should determine if it is feasible to permit this loan.

```python
X_train, y_train, X_test, y_test = load_data('loans')
best_model, results = automate_model_selection(X_train, y_train)
```

As we can see from the results, the best model is OCSVM - One Class Support Vector Machine


```python
plot_results_dict(best_model, results)
```

## Pulsar Detection
This dataset is a tabular dataset that contains measures of stars, some of them are a rare type of stars which called pulsars, and the goal is to detect them.

```python
X_train, y_train, X_test, y_test = load_data('pulsars')
best_model, results = automate_model_selection(X_train, y_train)
```

As we can see from the results, the best model is HBOS - Histogram-based Outlier Detection

```python
plot_results_dict(best_model, results)
```
